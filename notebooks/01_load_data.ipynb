```python
import pandas as pd

df = pd.read_csv("../data/habits.csv")
df
```

```python
df["date"] = pd.to_datetime(df["date"])
df.dtypes
```

```python
daily_progress = (
    df.groupby("date")["completed"]
      .mean()
      .reset_index(name="completion_rate")
)

daily_progress
```

```python
habit_progress = (
    df.groupby("habit_name")["completed"]
      .mean()
      .sort_values(ascending=False)
)

habit_progress
```
